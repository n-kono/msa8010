## String Slicing

`str_variable[start_position:end_position:step]`
- start_position: included
- end_position: excluded
- step: default is `1`




```python
mystr = "HelloWorld"
print(mystr) #The whole string
print(mystr[0:10]) #Start from 0, end just before index 10
print(mystr[0:len(mystr)]) #Similar to above, as len(mystr) is 10 in this example
print(mystr[0:]) #From 0, to the end (wherever it is)
print(mystr[:]) #Similar to above
```

```python
print(mystr[-1])  #Last item (last character)
print(mystr[-2:]) #Last two charachters 
print(mystr[:-2]) #The whole list except the last two characters
```

```python
print(mystr[::-1])    # all items in the array, reversed
print(mystr[1::-1])   # the first two items, reversed
print(mystr[:-3:-1])  # the last two items, reversed
print(mystr[-3::-1])  # everything except the last two items, reversed
```

```python
CODE = ""
s = "You can have data without information, but you cannot have information without data."

#In the code below, replace CODE with your code.

print(f"2.1 The length of s is {CODE}") #for example, here replace CODE with len(s)
print(f"2.2 The first occurrence of the letter 'a' is at index " + CODE) #s.index("a")


# Slicing the string into bits
print(f"2.3 The first five characters of s: {CODE}") #Should print You c
print(f"2.4 The last character is {CODE}") #Should print .
print(f"2.5 The last 5 characters are {CODE}") #Should print data.
print(f"2.6 The last 5 characters, without the tailing '.' are: {CODE}") #Should print data
print("The last five characters are '%s'" % s[-5:]) # 5th-from-last to end

# Convert everything to uppercase
print("String in uppercase: %s" % s.upper())

# Convert everything to lowercase
print("String in lowercase: %s" % s.lower())

# Check how a string starts
if s.startswith("You"):
    print("OK!")

# Check how a string ends
if s.endswith("data."):
    print("OK!")

# Split the string into three separate strings,
# each containing only a word
sp = s.split(" ")
print(f"Split the words of the string: {sp}")
```
